# adding up the elements from all of a nested lists:

```python
t = [ [ 1, 2 ] , [ 3 ], [ 4, 5, 6 ], 10]
total = 0
for nested in t:
    if type(nested) is list:  
        total += sum(nested)
    elif type(nested) is int:
        total += nested
print(total)
```

# Taking a list of numbers and returning the cumulative sum; that is, a new list where the ith element is the sum of the first i+1 elements from the original list.

```python
total = 0
res = []
t = [ 1, 2, 3 ]
for x in t:
    total += x
    res.append(total)
print(res)
```

# Check if there is any element in a given list that appears more than once.

```python
s = [1,4,2,10,89,21,1]
s.sort()
check = False
# check for adjacent elements that are equal
for i in range(len(s)-1):
    if s[i] == s[i+1]:
            check = True
print(check)
```

# split a list every n-th element.

```python
step = 3
C = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n']
A = []
for i in range(0,3):
    A.append(C[i::step])
#A = [C[i::step] for i in range(step)]
print(A)
```

# Write a Python program to get the largest number from a list.

```python
mylist = [56,3,23]
max_item = mylist[ 0 ]
for a in mylist:
    if a > max_item:
        max_item = a
print(max_item)
```

# Write a Python program to remove duplicates from a list

```python
a = [10,20,30,20,10,50,60,40,80,50,40]

uniq_items = []
for x in a:
    if x not in uniq_items:
        uniq_items.append(x)
        
print(uniq_items)
```

# Write a Python program that takes two lists and prints True if they have at least one common member.

```python
list1,list2 = [1,2,3,4,5], [5,6,7,8,9]
result = False
for x in list1:
    for y in list2:
        if x == y:
            result = True
print(result)
```
