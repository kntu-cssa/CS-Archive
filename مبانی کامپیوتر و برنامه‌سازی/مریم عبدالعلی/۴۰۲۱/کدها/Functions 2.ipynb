# Q1. Write a Python program to sum nested lists.

```python
def recursive_list_sum(seq):
	total = 0
	for element in seq:
		if isinstance(element, list):
			total = total + recursive_list_sum(element)
		else:
			total = total + element

	return total
print( recursive_list_sum([1, 2, [3,4],[5,6]]))
```

# Q2. Write a Python program to get the sum of a non-negative integer.

```python
def sumDigits(n):
    if n==0:
        return 0
    else:
        return n%10 + sumDigits(n//10)
sumDigits(5243)
```

# Q3. Tower of Hanoi



Main idea:

    - Shift ‘N-1’ disks from ‘A’ to ‘B’, using C.
    - Shift last disk from ‘A’ to ‘C’.
    - Shift ‘N-1’ disks from ‘B’ to ‘C’, using A.

```python
def TowerOfHanoi(n, from_rod, to_rod, aux_rod): 
    if n == 0: 
        return
    TowerOfHanoi(n-1, from_rod, aux_rod, to_rod) 
    print("Move disk", n, "from rod", from_rod, "to rod", to_rod) 
    TowerOfHanoi(n-1, aux_rod, to_rod, from_rod) 
  

N = 3
  
# A, C, B are the name of rods 
TowerOfHanoi(N, 'A', 'C', 'B') 
```

# Q4. Pascals Triangle

```python
def pascal(n):
    if n == 1:
        return [1]
    else:
        line = [1]
        previous_row = pascal(n-1)
        for i in range(len(previous_row)-1):
            line.append(previous_row[i]+previous_row[i+1])
        line +=[1]
        return line
pascal(3)
```

# Q5. Greatest Common Divisor (GCD) of 2 Numbers

```python
def gcd(a,b):
    if a==0:
        return b
    if b==0:
        return a
    return gcd(a%b, b) if a > b else gcd(a, b%a)

gcd(30, 16)
```
