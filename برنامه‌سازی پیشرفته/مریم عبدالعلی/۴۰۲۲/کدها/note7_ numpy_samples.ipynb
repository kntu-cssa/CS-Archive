##### Q2.  How to add a border (filled with 0's) around an existing array?

```python
Z = np.arange(0,25).reshape((5,5))
print(Z)

a = np.empty((Z.shape[0]+2,Z.shape[1]+2))
a[1:-1,1:-1]=Z[:,:]
a[:, [0, -1]] = 0
a[[0, -1], :] = 0
print(a)
# or use np.pad
```

##### Q3. Subtract the mean of each row of a matrix (be careful)

```python
X = np.arange(0,25).reshape((5,5))
X = X - X.mean(axis=1).reshape(-1,1)
X
```

Q4. How to sort an array by the 2nd column?

```python
X = np.random.randn(5,5)
print(X)
X=X[np.argsort(X[:,1], axis=0)]
print(X)
```

##### Q9. How to compute averages using a sliding window over an array?

```python
n = 3
Z = np.arange(1,11)
Z=np.insert(Z,0,0)
res = np.cumsum(Z, dtype=float)
conv = (res[n:]-res[:-n])/n
conv
```

##### Q9. Consider 2 sets of points P0,P1 describing lines (2d) and a point p, how to compute distance from p to each line i (P0[i],P1[i])?



```python
P1 = np.random.uniform(-10,10,(10,2))
P2 = np.random.uniform(-10,10,(10,2))
p  = np.random.uniform(-10,10,( 1,2))

D = np.sqrt(((P2-P1)**2).sum(axis=1))
N1 = (P2[:,0]-P1[:,0])*(p[0,1]-P1[:,1])
N2 = (p[0,0]-P1[:,0])*(P2[:,1]-P1[:,1])
N = np.abs(N1 - N2)
dist = N / D
print(dist)
```

##### Q10. Converting a color image to grayscale



```python
from matplotlib.pyplot import imread, imshow

a = imread("sample.jpg")
b = np.sum(a * np.array([0.2126, 0.7152, 0.0722]), axis =2)
imshow(b, cmap="gray")
```
