# Introduction to Recursion in Python

## What is Recursion?
Recursion is a technique in programming where a function calls itself in order to solve a problem. It is a fundamental concept in computer science and is widely used in algorithms such as tree traversals, sorting, and backtracking. 

Every recursive function must have two main components:
1. **Base case**: The condition that stops the recursion to prevent infinite loops.
2. **Recursive case**: The part where the function calls itself with a smaller problem.

Recursion is particularly useful when dealing with problems that can be broken down into smaller subproblems of the same type.

## Why Use Recursion?
- It simplifies complex problems by breaking them into smaller, more manageable problems.
- It is essential for problems that have a natural recursive structure, such as tree and graph algorithms.
- It makes code more readable and elegant compared to iterative solutions in some cases.

However, recursion has drawbacks:
- Excessive recursive calls can lead to a **stack overflow** if the recursion depth is too large.
- Recursive solutions may have more function call overhead compared to iterative solutions.


## Basic Example: Factorial Calculation

```python

def factorial(n):
    if n == 0:  # Base case
        return 1
    return n * factorial(n - 1)  # Recursive case

print("Factorial of 5:", factorial(5))  # Output: 120
```

## Fibonacci Sequence Using Recursion

```python

def fibonacci(n):
    if n <= 1:  # Base case
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)  # Recursive case

print("Fibonacci of 6:", fibonacci(6))  # Output: 8
```

## Recursion on Lists: Sum of Elements

```python

def sum_list(lst):
    if not lst:  # Base case: empty list
        return 0
    return lst[0] + sum_list(lst[1:])

print("Sum of [1, 2, 3, 4, 5]:", sum_list([1, 2, 3, 4, 5]))  # Output: 15
```

## String Manipulation: Reverse a String Using Recursion

```python

def reverse_string(s):
    if len(s) == 0:  # Base case
        return s
    return reverse_string(s[1:]) + s[0]

print("Reverse of 'hello':", reverse_string("hello"))  # Output: 'olleh'
```

## Generating All Permutations of a List Using Recursion

```python

def permutations(lst):
    if len(lst) == 0:
        return [[]]
    result = []
    for i in range(len(lst)):
        remaining_elements = lst[:i] + lst[i+1:]
        for p in permutations(remaining_elements):
            result.append([lst[i]] + p)
    return result

print("Permutations of [1,2,3]:", permutations([1, 2, 3]))
```

## Solving the Tower of Hanoi Problem

The Tower of Hanoi is a classic puzzle that involves moving a set of disks from one peg to another, following a set of rules. It's a great example of recursion in computer science.

Problem Description:
There are three pegs: A, B, and C.
There are n disks of different sizes, initially stacked in ascending order (smallest disk on top) on peg A.
The goal is to move all the disks from peg A to peg C, following these rules:
- You can only move one disk at a time.
- A disk can only be placed on top of a larger disk or on an empty peg.
- You can use peg B as an auxiliary peg to help with the moves.



```python

def tower_of_hanoi(n, source, auxiliary, target):
    if n == 1:
        print(f"Move disk 1 from {source} to {target}")
        return
    tower_of_hanoi(n - 1, source, target, auxiliary)
    print(f"Move disk {n} from {source} to {target}")
    tower_of_hanoi(n - 1, auxiliary, source, target)

tower_of_hanoi(3, 'A', 'B', 'C')
```

## Checking if a String is a Palindrome

```python

def is_palindrome(s):
    if len(s) <= 1:
        return True
    if s[0] != s[-1]:
        return False
    return is_palindrome(s[1:-1])

print("Is 'racecar' a palindrome?", is_palindrome("racecar"))  # Output: True
```

## Finding the Greatest Common Divisor (GCD)

```python

def gcd(a, b):
    if b == 0:
        return a
    return gcd(b, a % b)

print("GCD of 48 and 18:", gcd(48, 18))  # Output: 6
```

## Binary Search Using Recursion

Binary Search is an efficient algorithm for finding a target value within a sorted array or list. It works by repeatedly dividing the search interval in half, which makes it much faster than a linear search.


```python

def binary_search(arr, target, left, right):
    if left > right:
        return -1
    mid = (left + right) // 2
    if arr[mid] == target:
        return mid
    elif arr[mid] < target:
        return binary_search(arr, target, mid + 1, right)
    else:
        return binary_search(arr, target, left, mid - 1)

arr = [1, 2, 3, 4, 5, 6, 7, 8, 9]
print("Index of 5:", binary_search(arr, 5, 0, len(arr)-1))  # Output: 4
```

## Generating Subsets of a List

```python
def subsets(lst):
    if not lst:
        return [[]]
    rest_subsets = subsets(lst[1:])
    result = rest_subsets[:]
    for subset in rest_subsets:
        result.append([lst[0]] + subset)
    return result

print("Subsets of [1, 2, 3]:", subsets([1, 2, 3]))
```
