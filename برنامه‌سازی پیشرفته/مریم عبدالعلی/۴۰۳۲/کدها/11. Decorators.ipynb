### 1. Decorator for Timing Function Execution
This decorator measures the time taken by the decorated function to execute, which is useful for performance profiling.

```python
import time

def timeit(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        print(f"{func.__name__} executed in {end_time - start_time:.4f} seconds")
        return result
    return wrapper

@timeit
def slow_function():
    time.sleep(2)

slow_function()
```

### 2. Decorator for Caching Results (Memoization)
This decorator caches the results of expensive function calls to improve performance on repeated calls with the same arguments.

```python
def memoize(func):
    cache = {}
    def wrapper(*args):
        if args in cache:
            return cache[args]
        cache[args]= func(*args)
        return cache[args]
    return wrapper

@memoize
def expensive_computation(n):
    print("Computing...")
    return n * n

print(expensive_computation(4))
print(expensive_computation(4))
```

### 3. Decorator for Access Control
This example shows how to implement a decorator that checks if a user has the necessary permissions to access a certain function.

```python
def requires_permission(permission):
    def decorator(func):
        def wrapper(user, *args, **kwargs):
            if permission not in user["permissions"]:
                raise PermissionError(f"User does not have '{permission}' permission.")
            return func(user,*args, **kwargs)
        return wrapper
    return decorator
        

@requires_permission('admin')
def delete_user(user, user_id):
    print(f"User {user_id} has been deleted by {user['name']}.")

user = {'name': 'Alice', 'permissions': ['admin']}
delete_user(user, 123)  # Success

user_no_permission = {'name': 'Bob', 'permissions': ['user']}
delete_user(user_no_permission, 123)  # Raises PermissionError
```

### 4. Enforcing Type Checking
Ensures function arguments have specific types.

```python
def type_check(*types):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for arg,type in zip(args, types):
                if not isinstance(arg, type):
                    raise TypeError
            return func(*args, **kwargs)
        return wrapper
    return decorator

@type_check(int , int)
def multiply(a,b):
    return a*b

multiply(3,4)
```

### 5. Rate Limiting
Limits how often a function can be called.

```python
import time
def rate_limiter(interval):
    def decorator(func):
        last_call = [0]
        def wrapper(*args, **kwargs):
            elapsed = time.time()-last_call[0]
            if elapsed < interval :
                print(f"Rate limit exceeded. Try again in {interval - elapsed:.2f} seconds.")
                return
            last_call[0] = time.time()
            return func(*args, **kwargs)
        return wrapper
    return decorator

@rate_limiter(3)
def api_request():
    print("API request made.")

api_request()
time.sleep(2)
api_request()  # This will be rate-limited
time.sleep(3)
api_request()
```

### 6. Singleton Pattern
Ensures only one instance of a class is created.

```python
def singleton(cls):
    instances = {}
    def wrapper(*args, **kwargs):
        if cls not in instances:
            instances[cls] = cls(*args,**kwargs)
        return instances[cls]
    return wrapper

@singleton
class Sample:
    def __init__(self):
        print("Object created")

sample1 = Sample()
sample2 = Sample()
print(sample1 is sample2)
```
