# Basics of Networking

Networking is the process that enables communication and information sharing across devices. It involves the interconnection of two or more devices to allow them to exchange data. A collection of interconnected devices, whether computers, servers, or other devices, forms a network. Networks are fundamental to the functioning of the internet and local systems, and they serve as the backbone for applications such as web browsing, file sharing, email, and much more.

## Key Concepts

- **IP (Internet Protocol)**: An address system that uniquely identifies each device connected to a network. It ensures data is sent to the right location. There are two versions: IPv4 and IPv6.
- **Port**: A number used to identify a specific process or service on a device. It is essential for directing network traffic to the right application. Common ports include HTTP (80), HTTPS (443), and FTP (21).
- **Socket**: A software abstraction used to facilitate communication between two devices over a network. A socket is bound to a specific IP address and port, enabling data transfer.
- **Client/Server**: Two roles in a network communication. The server hosts resources or services and waits for requests from clients. A client connects to the server to access those services or resources.
- **TCP/IP (Transmission Control Protocol/Internet Protocol)**: A suite of communication protocols that enable reliable, ordered, and error-checked data transmission over a network. It is the foundation of internet and intranet communications.
- **UDP (User Datagram Protocol)**: A connectionless protocol that sends data without establishing a connection, offering lower overhead but no guarantees for data integrity or order.
- **Network Layers**: The different levels of communication that manage specific aspects of networking. The most well-known model is the OSI (Open Systems Interconnection) model, which has 7 layers ranging from physical connections to application-level protocols.

## Understanding Sockets

A socket is a fundamental concept in network programming. It serves as a communication endpoint for sending and receiving data over a network. Each socket is associated with an IP address and a port number, both of which help identify the source and destination for data transmission.

In more detail, a socket binds to an IP address and port on the local machine and provides a communication path to a remote system with its own IP address and port. This abstraction enables the communication between devices running on different machines or within the same network. 

In socket programming, you create socket objects within your code to facilitate network communication. You can perform various operations on these sockets, such as listening for incoming connections, sending data, receiving data, and closing the connection. These socket operations are performed using a variety of methods provided by the socket API.

Under the hood, these methods interact with your operating system’s kernel, specifically the network stack, which handles the actual transmission of data across the network. The kernel ensures that data is properly packaged, addressed, and routed between systems.


## Sockets and Client-Server Communication

Sockets are crucial in client-server communication. Sockets are connection endpoints used by clients and servers to exchange data. In client-server architecture, the server typically listens for incoming requests, while the client sends requests to the server.

- **Server Side**: A server listens for incoming connection requests on a specific port. Once a request is received, it establishes a connection through a new socket, processes the request, and sends back the appropriate response.
- **Client Side**: The client initiates a connection to the server using the server’s IP address and port. Once connected, it sends requests and waits for the server’s response.

This interaction follows a request-response model, where the client requests data or services, and the server provides them. Each client-server interaction involves establishing and maintaining socket connections to ensure that data flows in both directions.

## Python Socket Library

Python provides a built-in socket library that facilitates socket programming. The `socket` module in Python contains functions, constants, and classes that help create, manage, and operate on sockets.

Commonly used functions in the Python `socket` library include:

- **socket()**: Creates a new socket object. This is the first step in any socket communication.
- **bind()**: Associates the socket with a specific address (IP) and port, allowing it to listen for incoming connections on that address.
- **listen()**: Puts the socket into listening mode, awaiting connection requests. It is used only on the server side.
- **accept()**: Accepts an incoming connection request from a client. It returns a new socket object that is used for communication with the client.
- **connect()**: Establishes a connection to a remote server using its IP address and port. This is used on the client side.
- **send()**: Sends data through the socket. This method transmits data over an open connection.
- **recv()**: Receives data from the socket. It listens for incoming data from a remote connection.
- **close()**: Closes the socket connection, freeing the associated resources. This is crucial for cleanup once communication has ended.



## Example of Server-Client Communication

Here is a simple example of how server-client communication works using Python's `socket` library:


#### Sever.py Code

````python
import socket


def run_server():
    # create a socket object
    server = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

    server_ip = "127.0.0.1"
    port = 8000

    # bind the socket to a specific address and port
    server.bind((server_ip, port))
    # listen for incoming connections
    server.listen(0)
    print(f"Listening on {server_ip}:{port}")

    # accept incoming connections
    client_socket, client_address = server.accept()
    print(f"Accepted connection from {client_address[0]}:{client_address[1]}")

    # receive data from the client
    while True:
        request = client_socket.recv(1024)
        request = request.decode("utf-8") # convert bytes to string
        
        # if we receive "close" from the client, then we break
        # out of the loop and close the conneciton
        if request.lower() == "close":
            # send response to the client which acknowledges that the
            # connection should be closed and break out of the loop
            client_socket.send("closed".encode("utf-8"))
            break

        print(f"Received: {request}")

        response = "accepted".encode("utf-8") # convert string to bytes
        # convert and send accept response to the client
        client_socket.send(response)

    # close connection socket with the client
    client_socket.close()
    print("Connection to client closed")
    # close server socket
    server.close()


run_server()
````

#### Client Code

````python
import socket


def run_client():
    # create a socket object
    client = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

    server_ip = "127.0.0.1"  # replace with the server's IP address
    server_port = 8000  # replace with the server's port number
    # establish connection with server
    client.connect((server_ip, server_port))

    while True:
        # input message and send it to the server
        msg = input("Enter message: ")
        client.send(msg.encode("utf-8")[:1024])

        # receive message from the server
        response = client.recv(1024)
        response = response.decode("utf-8")

        # if server sent us "closed" in the payload, we break out of the loop and close our socket
        if response.lower() == "closed":
            break

        print(f"Received: {response}")

    # close client socket (connection to the server)
    client.close()
    print("Connection to server closed")

run_client()
````

### Multithreaded server
In the previous example, we observed how a server responds to requests from a single client. However, in real-world applications, servers often need to handle multiple clients simultaneously. This is where multithreading becomes essential. Multithreading allows a program to execute multiple tasks concurrently, enabling the server to manage multiple client connections at the same time.

A thread is a lightweight unit of execution within a process. Unlike processes, which operate independently and require significant system resources, threads share the same memory space within a process, making them more efficient for handling multiple tasks concurrently.

When implementing a multi-client server, the idea is to spawn a new thread for each incoming client connection. Each thread runs independently, allowing the server to continue accepting new connections while simultaneously processing data from existing clients. This ensures smooth and responsive communication without blocking other clients from connecting.

By leveraging multithreading, the server can efficiently manage multiple clients without significant overhead, improving scalability and performance. However, handling multiple threads also introduces challenges such as synchronization issues, race conditions, and potential resource contention, which need to be carefully managed using thread synchronization techniques like locks, semaphores, and thread-safe data structures.

#### Multithreading server code

This Python script implements a multi-client TCP server using the socket and threading modules. The server listens on 127.0.0.1:8000 and accepts multiple client connections, each handled in a separate thread. When a client connects, the server starts a new thread running handle_client(), which continuously listens for messages from the client. If the client sends a message, the server prints it and responds with "accepted"; if the client sends "close", the server replies with "closed" and terminates the connection. The use of threads allows multiple clients to interact with the server simultaneously without blocking new connections. The server runs indefinitely, continuously accepting and managing client connections until manually stopped.

````python
import socket
import threading


def handle_client(client_socket, addr):
    try:
        while True:
            # receive and print client messages
            request = client_socket.recv(1024).decode("utf-8")
            if request.lower() == "close":
                client_socket.send("closed".encode("utf-8"))
                break
            print(f"Received: {request}")
            # convert and send accept response to the client
            response = "accepted"
            client_socket.send(response.encode("utf-8"))
    except Exception as e:
        print(f"Error when hanlding client: {e}")
    finally:
        client_socket.close()
        print(f"Connection to client ({addr[0]}:{addr[1]}) closed")


def run_server():
    server_ip = "127.0.0.1"  # server hostname or IP address
    port = 8000  # server port number
    # create a socket object
    try:
        server = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        # bind the socket to the host and port
        server.bind((server_ip, port))
        # listen for incoming connections
        server.listen()
        print(f"Listening on {server_ip}:{port}")

        while True:
            # accept a client connection
            client_socket, addr = server.accept()
            print(f"Accepted connection from {addr[0]}:{addr[1]}")
            # start a new thread to handle the client
            thread = threading.Thread(target=handle_client, args=(client_socket, addr,))
            thread.start()
    except Exception as e:
        print(f"Error: {e}")
    finally:
        server.close()


run_server()
````
